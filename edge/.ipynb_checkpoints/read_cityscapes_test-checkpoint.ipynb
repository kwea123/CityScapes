```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import cv2
import json
```

```python
img_name = 'darmstadt_000000_000019'
city = img_name.split('_')[0]
left_img = np.array(Image.open('/home/ubuntu/data/CityScapes/leftImg8bit_trainvaltest/leftImg8bit/train/%s/%s.png'%(city, img_name+'_leftImg8bit')))
ss = np.array(Image.open('/home/ubuntu/data/CityScapes/gtFine_trainvaltest/gtFine/train/%s/%s.png'%(city, img_name+'_gtFine_color')))[:, :, :3]

with open('/home/ubuntu/data/CityScapes/gtFine_trainvaltest/gtFine/train/%s/%s.json'%(city, img_name+'_gtFine_polygons')) as f:
    label = json.load(f)
```

```python
plt.figure(figsize=(20, 10))
plt.imshow(left_img)
plt.figure(figsize=(20, 10))
plt.imshow(ss)
```

```python
def create_edge_map(label, ss):
    ss_z = np.zeros_like(ss)
    for obj in label['objects']:
        arr = np.array(obj['polygon'])
        cv2.fillPoly(ss_z, [arr.reshape(-1,1,2)], (0, 0, 0))
        cv2.polylines(ss_z, [arr.reshape(-1,1,2)], True, (255,255,255))
    return ss_z
```

```python
plt.figure(figsize=(20, 10))
edge = create_edge_map(label, ss)
plt.imshow(edge, cmap='gray')
```

## Generate all edges

```python
import os
split = 'val'
cities = os.listdir('/home/ubuntu/data/CityScapes/leftImg8bit_trainvaltest/leftImg8bit/%s'%split)
for city in cities:
    if not os.path.isdir('/home/ubuntu/data/CityScapes/edge_trainvaltest/%s/%s'%(split, city)):
        os.mkdir('/home/ubuntu/data/CityScapes/edge_trainvaltest/%s/%s'%(split, city))
    img_names = os.listdir('/home/ubuntu/data/CityScapes/leftImg8bit_trainvaltest/leftImg8bit/%s/%s'%(split, city))
    for img_name in img_names:
        with open('/home/ubuntu/data/CityScapes/gtFine_trainvaltest/gtFine/%s/%s/%s.json'%(split, city, img_name[:-16]+'_gtFine_polygons')) as f:
            label = json.load(f)
        ss = np.array(Image.open('/home/ubuntu/data/CityScapes/gtFine_trainvaltest/gtFine/%s/%s/%s.png'%(split, city, img_name[:-16]+'_gtFine_color')))[:, :, :3]
        edge = create_edge_map(label, ss)
        plt.imsave('/home/ubuntu/data/CityScapes/edge_trainvaltest/%s/%s/%s.png'%(split, city, img_name[:-16]+'_edge'), edge, cmap='gray')
```

## Generate train pairs txt

```python
import os
lines = []
for split in ['train', 'val']:
    cities = os.listdir('/home/ubuntu/data/CityScapes/leftImg8bit_trainvaltest/leftImg8bit/%s'%split)
    for city in cities:
        img_names = os.listdir('/home/ubuntu/data/CityScapes/leftImg8bit_trainvaltest/leftImg8bit/%s/%s'%(split, city))
        for img_name in img_names:
            img_path = 'leftImg8bit_trainvaltest/leftImg8bit/%s/%s/%s'%(split, city, img_name)
            edge_path = 'edge_trainvaltest/%s/%s/%s.png'%(split, city, img_name[:-16]+'_edge')
            lines += [img_path+' '+edge_path+'\n']

with open('/home/ubuntu/data/CityScapes/train_edge_pairs.txt', 'w') as f:
    f.writelines(lines)
```
